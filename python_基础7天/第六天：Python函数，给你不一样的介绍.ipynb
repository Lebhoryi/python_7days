```python
'''
文件名字：第六天：Python函数，给你不一样的介绍.ipynb
摘要：    项目实战:如何利用 python 删除一个指定的文件

版本：V0.0.1

作者：Lebhoeryi@gmail.com
时间：20181010

'''


#################################################################
# 笔记：
# 传递字典
#    函数中传递字典。在 Python 中,使用**作为形参接受实参传递的键值对。注意区别:
# 在形参列表中,*表示传递空元组、**表示传递字典。
#
# 项目要求:
#     必须使用函数，删除一个指定的文件
# 提示与问题拆解
#     a) 如何找到指定路径下的所有文件
#     b) 如何找到指定的文件名在指定的路径下
#     c) 如何删除找到的文件为新的名字
#     d) 使用 os 模块下的 remove()方法
# 
# 解题思路（思路很重要，我的思路是第三点倒推的，正推想了很久）：
#     1.没有文件就创建一个文件，函数cre_file
#     2.删除文件，函数del_file
#     3.del_file:
#         a.传进来的是带路径带文件名的参数
#         b.获取文件的绝对路径，包含文件名字的路径
#         c.分离文件名和不包含文件名的路径
#         d.枚举不包含文件名的路径下的所有文件
#         e.判断文件名是否为要删除的文件，是就删除，否就跳过
#     4.主函数
#
##################################################################
import os

def cre_file(file_name):
    """生成一个新的文件
    """
    with open(file_name,"w") as f:
        f.write("这是一个测试文件")
    print(f"成功创建新的文件：{file_name}!")

def del_file(file_name):
    '''定义一个删除指定文件的函数
    
    Args:
        file_path: 要删除的文件路径
        del_file_name: 要删除的文件名  
    '''
    
    # 获取文件的绝对路径
    file_path = os.path.abspath(file_name)
#     print(file_path)

    # 返回一个路径的目录名和文件名
    file_path_del = os.path.split(file_path)[0]
    file_name_del = os.path.split(file_path)[1]
#     print(f"file_path_del: {file_path_del}")
#     print(f"file_name_del: {file_name_del}")
    
    # 枚举不带文件名字的路径下的所有路径，目录，文件
    for root, dirs, files in os.walk(file_path_del):
        for file in files:
            if file == file_name_del:
                os.remove(file_name_del)
                print(f"删除文件：{del_file_name}成功")
                

if __name__ == "__main__":
    file_name = "./test.txt"   # 传入的参数，带名字的路径
    
    # 判断文件是否存在，是就删除，否就先创建再删除
    if os.path.exists(del_file_name):
        del_file(del_file_name)

    else:
        print(f"没有这个文件：{del_file_name},即将创建...")
        cre_file(del_file_name)
        del_file(del_file_name)
    
```
