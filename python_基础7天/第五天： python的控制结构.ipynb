```python
'''
文件名：第五天： python的控制结构.ipynb 
摘要： 项目实战 1--生成九九乘法表 

版本：V0.0.1

作者：Lebhoryi@gmail.com
时间：20181007
'''

#############################################################
# 项目描述：根据嵌套 for 循环生成九九乘法表
# 提示：
#     根据九九乘法表的规律，找到两个 for 循环的边界值 。
#     使用 range()函数生成数字集合
#############################################################

# i 为右边界，每行增加，同行不变
for i in range(1,10):
    # j 为左边界，按列增加，每行比上一行多一个数
    for j in range(1,i+1):
        # python3 不换行：print(,end="")
        print('{}*{}    '.format(j,i),end="")
        # 换行条件：左边值等于右边值
        if j == i:
            print("\n")
```

```python
#############################################################
# *
# ***
# *****
# *******
# *****
# ***
# *
#
# 项目实战 2--打印图 
# 项目描述：使用 for循环打印上图
# 提示：
#     使用两个 for循环
#     找对 in后面的边界值，使用 range()
#############################################################

# 打印上四行
for i in range(1,8,2):
    for j in range(i):
        print("*", end="")
    print("\n")
# 打印下三行
for i in range(5,0,-2):
    for j in range(i):
        print("*", end="")
    print("\n")

# 简介版
# 打印上四行
for i in range(1,8,2):
    print("*" * i)
    print("\n")
# 打印下三行
for i in range(5,0,-2):
    print("*" * i)
    print("\n")
```

```python
#############################################################
# 项目实战 3-创建自己的通讯录 
# 项目描述： 请为自己创建一个通讯录，效果如下图所示。要求，必须有欢迎界面和指令
# Python 实战圈出品
# 
# 提示
# |---欢迎进入通讯录程序---|
# |---1、 查询联系人资料---|
# |---2、 插入新的联系人---|
# |---3、 删除已有联系人---|
# |---4、 退出通讯录程序---|
# 请输入指令代码：1
# 请输入联系人姓名: 刘德华
# 该联系人不存在！
# 请输入指令代码：2
# 请输入联系人姓名:刘德华
# 请输入联系人电话：136XXXXXXX
# 联系人加入成功！
# 请输入指令代码：2
# 请输入联系人姓名:杨幂
# 请输入联系人电话：177xxxxxxxxx
# 联系人加入成功！
# 请输入指令代码：1
# 请输入联系人姓名:杨幂
# 杨幂 : 177xxxxxxxxx
# 请输入指令代码：4
# |---感谢使用通讯录程序---|
# 提示：
#     使用字典构建通讯录
#     使用 while 循环一直判断是不是进行
#     使用 if 语句判断指令类型
#############################################################

import time
contacts = {}    # 通讯录字典
flag = True    # while循环判断条件

while flag:
    print("|---欢迎进入通讯录程序---|")
    print("|---1、 查询联系人资料---|")
    print("|---2、 插入新的联系人---|")
    print("|---3、 删除已有联系人---|")
    print("|---4、 退出通讯录程序---|")
    # 输入指令序号
    num = int(input("请输入指令代码："))
        
    # 查询联系人资料
    if num == 1:
        # 输入查询名字
        old_name = input("请输入联系人姓名: ")
        # 判断名字是否在字典里面
        if old_name in contacts.keys():
            # 打印该联系人
            print("{}的联系方式为：{}".format(old_name, contacts[old_name]))
        else:
            print("该联系人不存在！")
            # 跳出来的的时间太长，休眠5s
            time.sleep(5)
  
    # 插入新的联系人
    elif num == 2:
        # 输入新的联系人名字
        new_name = input("请输入联系人姓名:")
        # 输入电话号码
        new_phone_num = str(input("请输入联系人电话："))
        # 名字不为空且电话号码为11位，录入成功
        if new_name and len(new_phone_num) == 11:
            # 字典加键值对，名字：联系电话
            contacts[new_name] = new_phone_num
            print("联系人加入成功！")
        
    # 删除已有联系人
    elif num == 3:
        # 输入删除的联系人名称
        del_name = input("请输入要删除的联系人：")
        # 如果删除的联系人名称在字典里面就删除
        if del_name in contacts.keys():
            # 字典删除键值对方法
            contacts.pop(del_name)
            print("删除成功")
        else:
            print("该联系人不存在！")
            
    # 退出通讯录程序
    elif num == 4:
        print("|---感谢使用通讯录程序---|")
        break
    
    else:
        print("输入的指令有误，请重新输入....")
        continue
```
